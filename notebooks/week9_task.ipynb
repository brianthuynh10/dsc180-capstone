```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/week9.csv')
df['Tags'] = df['Tags'].apply(lambda x: int(x[-2:]))
df = df[['Tags', 'test_r', 'test_mae']]
df.head()
```

```python
summary = df.groupby('Tags')['test_r'].agg(['mean', 'std']).reset_index()
```

```python
# Plot
plt.figure(figsize=(8,5))
plt.errorbar(
    summary['Tags'],
    summary['mean'],
    yerr=summary['std'],
    fmt='o-',            # line + points
    capsize=5,           # little cross bar at top of error bar
    ecolor='red',  # color of error bar
)

# make x-ticks 11, 13, 16, 19
plt.xticks([11, 13, 16, 19])
plt.xlabel("VGG Model Depth")
plt.ylabel("Test Pearson r")
plt.title("Test r vs VGG Model Depth (with error bars across seeds)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
